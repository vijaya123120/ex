```python
#Convert the string representations to actual lists of dictionaries
import ast
df['genres'] = df['genres'].apply(ast.literal_eval)

# Extract the genre names using a lambda function
names = df['genres'].apply(lambda x: [genre['name'] for genre in x])
```

```python
df['genres']=names
```

```python

# 'production_companies', 'production_countries', 'keywords', 'spoken_languages' columns contain string representations of lists of dictionaries

# Convert the string representations to actual lists of dictionaries
df['production_companies'] = df['production_companies'].apply(ast.literal_eval)
df['production_countries'] = df['production_countries'].apply(ast.literal_eval)
df['keywords'] = df['keywords'].apply(ast.literal_eval)
df['spoken_languages'] = df['spoken_languages'].apply(ast.literal_eval)

# Extract the values from the dictionaries in the 'production_companies' column
df['production_companies'] = df['production_companies'].apply(lambda x: [item['name'] for item in x])

# Extract the values from the dictionaries in the 'production_countries' column
df['production_countries'] = df['production_countries'].apply(lambda x: [item['name'] for item in x])

# Extract the values from the dictionaries in the 'keywords' column
df['keywords'] = df['keywords'].apply(lambda x: [item['name'] for item in x])

# Extract the values from the dictionaries in the 'spoken_languages' column
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: [item['name'] for item in x])


```

```python
import statistics as st
mean = []
median = []
mode = []

# Iterate over each column in the DataFrame
for col in DF[['budget', 'revenue', 'runtime']]:
    median.append(DF[col].median())         
    mean.append(DF[col].mean())            
    mode.append(st.mode(DF[col]))           

row_head = ['median', 'mean', 'mode']
col_names = ['budget', 'revenue', 'runtime']



d_data=pd.DataFrame([mean,median,mode],columns=col_names)
d_data.insert(0,'Measures',row_head)
d_data
```
