# UNIVARAITE ANALYSIS

```python
terr_df['Attack_Type'].value_counts()
```

```python
attack_type_counts = terr_df['Attack_Type'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
attack_type_counts.plot(kind='bar')
plt.title('Attack Types',fontweight='bold',fontsize=16)
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#50% Attacks happened throughBombing/Explosion
```

# TOP 10 COUNTRIES WHICH SUFFER MOST FROM TERRORISM

```python
plt.figure(figsize=(12,6))
sns.barplot(terr_df['Country'].value_counts()[:10].index,terr_df['Country'].value_counts()[:10].values,palette='viridis')
plt.title("Top 10 Countries Affected",fontweight='bold',fontsize=16)
plt.xlabel('country')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.show()
```

```python
(terr_df['Target_type'].value_counts().head(10)/terr_df['Target_type'].shape[0])*100
```

# Top 10 Attacked Targets

```python
plt.figure(figsize=(12,6))
sns.barplot(terr_df['Target_type'].value_counts()[:10].index,terr_df['Target_type'].value_counts()[:10].values,palette='magma')
plt.title("Attacked Targets",fontweight='bold',fontsize=16)
plt.xlabel('Target')
plt.ylabel('count')
plt.xticks(rotation=80)
plt.show()
```

# Most attacked regions and states

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

sns.barplot(terr_df['Region'].value_counts()[:10].index, terr_df['Region'].value_counts()[:10].values, palette='flare', ax=ax[0])
ax[0].set_title("Most Attacked Regions", fontsize=20,fontweight='bold')
ax[0].set_xlabel('Region')
ax[0].set_ylabel('Count of Attacks')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=90)

sns.barplot(terr_df['State'].value_counts()[:10].index, terr_df['State'].value_counts()[:10].values, palette='viridis', ax=ax[1])
ax[1].set_title("Most Attacked States", fontsize=20,fontweight='bold')
ax[1].set_xlabel('State')
ax[1].set_ylabel('Count of Attacks')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=80)

plt.show()
```

```python
#most Attacked region is Midddle East & North Africa . And Baghdad state is the most attacked one
```

```python
plt.figure(figsize=(12,6))
sns.barplot(terr_df['City'].value_counts()[1:10].index,terr_df['City'].value_counts()[1:10].values,palette='viridis')
plt.title("Top 10 Cities Affected",fontweight='bold',fontsize=20)
plt.xlabel('city')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Baghdaad city is the most affected 
```

```python

top_weapon_types = terr_df['Weapon_Type'].value_counts().head(5)
plt.figure(figsize=(6, 8))
plt.pie(top_weapon_types, labels=top_weapon_types.index, autopct='%1.1f%%', startangle=90)
plt.title('Weapon Types',fontweight='bold',fontsize=20)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python

plt.figure(figsize=(12,7))
x_year = terr_df['Year'].unique()
y_count_years=terr_df['Year'].value_counts(dropna=False).sort_index()
sns.barplot(x=x_year, y=y_count_years, palette='dark:salmon_r')
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=15)
plt.title('Attacks by Year',fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
terr_df.groupby(['Year'])['killed'].sum().plot(kind='bar',colormap='summer')
plt.title('Deaths in Different years',fontsize=20)
plt.xlabel('Years',fontsize=20)
plt.ylabel('Number of Deaths',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(y=terr_df['Group'].value_counts()[1:11].index,x=terr_df['Group'].value_counts()[1:11].values,palette='twilight_shifted')
plt.title("Top 10 Terrorist Groups",fontsize=20,fontweight='bold')
plt.xlabel('Number of Attacks',fontsize=15)
plt.ylabel('Terrorist Groups',fontsize=15)
plt.xticks(rotation=80)
plt.show()
```

# BIVARIATE ANALYSIS

```python


fig, ax = plt.subplots(1, 2, figsize=(12, 6))

terr_df.groupby('Region')['killed'].sum().sort_values(ascending=False).plot(kind='bar', colormap='Accent', ax=ax[0])
ax[0].set_title("Regions with Killed People", fontsize=20)
ax[0].set_xlabel('Region')
ax[0].set_ylabel('Number of Killed')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=90)

terr_df.groupby('Region')['wounded'].sum().sort_values(ascending=False).plot(kind='bar', colormap='Accent', ax=ax[1])
ax[1].set_title("Regions with Wounded People", fontsize=20)
ax[1].set_xlabel('Region')
ax[1].set_ylabel('Number of Wounded')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=90)

plt.show()
```

```python


fig, ax = plt.subplots(1, 2, figsize=(12, 6))

terr_df.groupby('Country')['killed'].sum().sort_values(ascending=False).head(10).plot(kind='bar', colormap='winter', ax=ax[0])
ax[0].set_title("Top 10 Countries with Highest Number of Killed People", fontsize=12)
ax[0].set_xlabel('Country')
ax[0].set_ylabel('Number of Killed')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=90)

terr_df.groupby('Country')['wounded'].sum().sort_values(ascending=False).head(10).plot(kind='bar', colormap='winter', ax=ax[1])
ax[1].set_title("Top 10 Countries with Highest Number of Wounded People", fontsize=12)
ax[1].set_xlabel('Country')
ax[1].set_ylabel('Number of Wounded')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=90)

plt.show()

```

```python
terr_df_taliban=terr_df[terr_df.Group=='Taliban']
```

```python
plt.figure(figsize=(13,7))
sns.barplot(terr_df_taliban['Year'].value_counts().index,terr_df_taliban['Year'].value_counts().values,palette='viridis')
plt.title('Terror Attack over the years by Taliban',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(13,7))
terr_df_taliban.groupby(['Year'])['killed'].sum().plot(kind='bar',colormap='RdBu')
plt.title('People killed by Taliban over the years',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of people killed',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

# Multivariate Analysis

```python
terr_df_2014=terr_df[terr_df.Year==2014]
```

```python

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

terr_df_2014.groupby('Country')['killed'].sum().sort_values(ascending=False).head(10).plot(kind='bar', colormap='RdBu', ax=ax[0])
ax[0].set_title("Top 10 Countries with Highest Number of Killed People in 2014", fontsize=10)
ax[0].set_xlabel('Country')
ax[0].set_ylabel('Number of Killed')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=90)

terr_df_2014.groupby('Country')['wounded'].sum().sort_values(ascending=False).head(10).plot(kind='bar', colormap='spring', ax=ax[1])
ax[1].set_title("Top 10 Countries with Highest Number of Wounded People in 2014", fontsize=10)
ax[1].set_xlabel('Country')
ax[1].set_ylabel('Number of Wounded')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=90)

plt.show()


```

```python
fig,ax=plt.subplots(1,2,figsize=(12,6))

# Regions that were attacked in 2014
plt.subplot(1, 2, 1)
sns.barplot(terr_df_2014['Region'].value_counts().index, terr_df_2014['Region'].value_counts().values, palette='magma')
plt.title('Most Attacked Regions in 2014', fontsize=15)
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Number of Attacks', fontsize=15)
plt.xticks(rotation=90)

# Top 10 Countries that were most attacked in 2014
plt.subplot(1, 2, 2)
sns.barplot(terr_df_2014['Country'].value_counts().head(10).index, terr_df_2014['Country'].value_counts().head(10).values, palette='magma')
plt.title('Top 10 Most Attacked Countries in 2014', fontsize=15)
plt.xlabel('Country', fontsize=15)
plt.ylabel('Number of Attacks', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```
